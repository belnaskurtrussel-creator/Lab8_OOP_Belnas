```python
class Laptop:
    def __init__(self, brand, processor, ram):
        self.brand = brand
        self.processor = processor
        self.ram = ram

    def get_specs(self):
        return f"Brand: {self.brand}, Processor: {self.processor}, RAM: {self.ram}GB"

my_laptop = Laptop("Dell", "Intel Core i7", 16)
print(my_laptop.get_specs())
```

```python
class StudentRecord:
    def __init__(self, student_name):
        self.student_name = student_name
        self.__grades = []   # Private attribute

    def add_grade(self, grade):
        self.__grades.append(grade)
        return f"Grade {grade} added."

    # Method to retrieve all grades
    def get_grades(self):
        return f"Grades: {self.__grades}"

student = StudentRecord("Kurt")

print(student.add_grade(90))
print(student.add_grade(85))
print(student.add_grade(92))

print(student.get_grades())
```

```python
class Device:
    def __init__(self, device_name):
        self.device_name = device_name

    def power_on(self):
        return f"{self.device_name} is now powered on."

class Smartphone(Device):
    def power_on(self):
        return f"{self.device_name} is booting up... Welcome to your smartphone!"

device = Device("Generic Device")
smartphone = Smartphone("Samsung Galaxy S23")

print(device.power_on())
print(smartphone.power_on())
```

```python
class Device:
    def __init__(self, device_name):
        self.device_name = device_name

    def power_on(self):
        return f"{self.device_name} is now powered on."

class Smartphone(Device):
    def power_on(self):
        return f"{self.device_name} is booting up... Welcome to your smartphone!"

class Laptop(Device):
    def power_on(self):
        return f"{self.device_name} is starting... Loading your desktop environment!"

def start_device(device):
    print(device.power_on())

phone = Smartphone("Samsung Galaxy S23")
laptop = Laptop("Dell XPS 15")

start_device(phone)
start_device(laptop)
```

```python
class TemperatureConverter:
    @staticmethod
    def celsius_to_fahrenheit(celsius):
        return (celsius * 9/5) + 32

celsius_input = float(input("Enter temperature in Celsius: "))

fahrenheit_result = TemperatureConverter.celsius_to_fahrenheit(celsius_input)

print(f"{celsius_input}°C is equal to {fahrenheit_result}°F")
```

```python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __sub__(self, other):
        return Vector(self.x - other.x, self.y - other.y)

    def __str__(self):
        return f"({self.x}, {self.y})"

v1 = Vector(3, 4)
v2 = Vector(1, 2)

print("Vector 1:", v1)
print("Vector 2:", v2)
print("Addition:", v1 + v2)
print("Subtraction:", v1 - v2)
```

```python
from abc import ABC, abstractmethod

class Appliance(ABC):
    @abstractmethod
    def turn_on(self):
        pass

class Refrigerator(Appliance):
    def turn_on(self):
        return "Refrigerator is now cooling."

fridge = Refrigerator()
print(fridge.turn_on())
```

```python
class A:
    def show(self):
        return "Class A"

class B(A):
    def show(self):
        return "Class B"

class C(A):
    def show(self):
        return "Class C"

class D(B, C):
    def show(self):
        return "Class D"

obj = D()
print(obj.show())

print("MRO of class D:", D.mro())
```
